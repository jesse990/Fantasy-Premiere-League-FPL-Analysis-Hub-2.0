```python
import pandas as pd
import requests
import os
```

```python
#url for all calls
base_url = "https://fantasy.premierleague.com/api/"
#folder where data will be saved       
folder = r"C:\Users\JesseOnu\FPL new\playerid" 
output = f"{folder}.csv"
```

```python
#call api
response = requests.get(base_url + "bootstrap-static/")
```

```python
#convert to json
data = response.json()
```

```python
print(type(data))
print(type(response))
print(response.status_code)
```

```python
#displaying availabl fields
print(data["elements"][0].keys())
```

```python
#keeping necesarry fields
df = pd.DataFrame(data["elements"])

df = df[["id", "code", "first_name", "second_name", "web_name", "photo", "element_type"]]
    
```

```python
df.head()
```

```python
#replace 'jpg' with 'png' as images dont work with jpg
df['photo'] = df['photo'].str.replace('jpg', 'png', regex = False)
```

```python
#add prefix to photo
df['photo'] = "https://resources.premierleague.com/premierleague25/photos/players/110x140/" + df['photo']
df.head(1)
```
